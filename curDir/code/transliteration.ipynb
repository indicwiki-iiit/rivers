```python
from anuvaad import Anuvaad
anu = Anuvaad("english-telugu")
```

```python
import pandas as pd
data = pd.read_excel('rivernames.xlsx')
                  
data
```

```python
for i in range(5):
    print(data.iloc[i,1])
```

```python
riverl=[]
for i in range(10000,15000):
    try:
        translated = anu.anuvaad(data.iloc[i,1])
        riverl.append(translated)
        print(data.iloc[i,1],translated)
    except:
        riverl.append("nan")

s=pd.DataFrame(riverl)
```

```python
s.to_excel('rivername tel 10కే_15k .xlsx',index=False)
```
