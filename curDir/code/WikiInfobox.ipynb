```python
df = pd.DataFrame()
blah = pd.read_excel('WIKIriverlinks.xlsx')
url_list = list(blah['Column1'].to_numpy())[1:2007]
# print(url_list)
for a in url_list:
    # URL = "https://en.wikipedia.org/wiki/Anadyr_(river)"
    if '\n' in a:
        a = a[:-1]
    URL = a
    print(URL)
    response = requests.get(URL)
    soup = bs4.BeautifulSoup(response.text, 'html')
    infobox = soup.find('table', {'class': 'infobox'})
    # for i in infobox:
    #     title = i.find('th', {'class': 'infobox-above'}).text
    s = soup.find_all('th', {'class': 'infobox-label'})
    head_list = []
    list_ = []
    listall = []
    
    i = 0
    while i<len(s):
        if('•' not in s[i].text):
            head_list.append(s[i].text)
            listall.append(head_list[-1])
            i += 1
        else:
            while i<len(s) and '•' in s[i].text:
                list_.append(head_list[-1]+'-'+s[i].text[3:])
                listall.append(list_[-1])
                i += 1
    val=soup.find_all('td', {'class': 'infobox-data'})
    d = {'rivername':[a.split('/')[-1]]}
    for i in range(len(val)):
        temp = normalize("NFKD", val[i].text)
        if 'coordinates' in listall[i].lower():
            if "ufeff" in temp:
                a = temp.split('/ \ufeff')[1].split('\ufeff')[0]
            temp = a
        if temp !=' ':
            d[listall[i]] = [temp]
        else:
            temp = ''
            d[listall[i]] = 'NaN'
    df = df.append(d, ignore_index=True)
df
```

```python
df.to_csv('infobox1_2006.csv', index=False)
```
